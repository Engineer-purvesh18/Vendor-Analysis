EXPLORATORY DATA ANALYSIS (VISUALIZATION)

```python
# Importing required libraries and packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sqlite3
from scipy.stats import ttest_ind
import scipy.stats as stats
warnings.filterwarnings('ignore')
```

```python
# Loading dataset
conn = sqlite3.connect('inventory.db') 
'''Creating connection with database (inventory.db)'''

# fetching vendor summary data and storing it in dataframe
df = pd.read_sql_query("select * from vendor_sales_summary",conn)
df.head()
```

```python
df.to_csv('vendor_sales_summary.csv',index  = False)
```

EXPLORING DATA THROUGH VISUALIZATIONS

```python
# Summary statistics for numerical columns
summary_stats = df.describe().T  
display(summary_stats)  
```

```python
# Mode for each numerical column
mode_values = df.mode().iloc[0]
print("\nMode Values:\n\n", mode_values)
```

```python
# Distribution Plots for Numerical Columns
numerical_cols = df.select_dtypes(include=np.number).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 4, i+1)  # Adjust grid layout as needed
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Outlier Detection with Boxplots
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 4, i+1)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```
